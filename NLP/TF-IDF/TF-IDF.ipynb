# TF-IDF

```python
import pandas as pd
df=pd.read_csv('./SMSSpamCollection.csv',
                    sep='\t',names=["label","message"])
```

```python
df
```

```python
import re
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
wordlemmatize=WordNetLemmatizer()
```

```python
corpus=[]
for i in range(0,len(df)):
    review=re.sub('[^a-zA-z]',' ',df['message'][i])
    review=review.lower()
    review=review.split()
    review=[wordlemmatize.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

## Create TF-IDF And NGrams

```python
from sklearn.feature_extraction.text importTfidfVectorizer
```

```python
tfidf=TfidfVectorizer(max_features=100)
```

```python
X=tfidf.fit_transform(corpus).toarray()
```

```python
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000, 
    formatter=dict(float=lambda x: "%.3g" % x))
```

```python
tfidf=TfidfVectorizer(max_features=100,ngram_range=(2,2))
X=tfidf.fit_transform(corpus).toarray()
```
