## Tokenization in NLP

```python
import nltk
nltk.download('punkt_tab')
nltk.download('punkt')
print('Done')
```

```python
corpus="""Hey there! Myself Dhyan Shah. I am CSE student at GCET.
I am AI/ML enthusiastspecializing in GenAI,AgenticAI, RAG , LLMs etc.
Connect with me !!
"""
```

```python
print(corpus)
```

```python
## Paragraph->Sentencte
from nltk.tokenize import sent_tokenize
documents=sent_tokenize(corpus)
```

```python
for sentence in documents:
    print(sentence)
```

```python
## paragraph -> words
## sentence -> words
from nltk.tokenize import word_tokenize
word_tokenize(corpus)
```

```python
for sentence in documents:
    print(word_tokenize(sentence))
```

```python
## also tokenizes punctuations
from nltk.tokenize import wordpunct_tokenize
wordpunct_tokenize(corpus)
```

```python
## doesnttokenizefullstopexceptlast
from nltk.tokenize import TreebankWordTokenizer
tokenizer=TreebankWordTokenizer()
tokenizer.tokenize(corpus)
```
