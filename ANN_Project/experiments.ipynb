```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder,OneHotEncoder
import pickle
print('Imported')
```

```python
## Loading the dataset
df=pd.read_csv('./Churn_Modelling.csv')
```

```python
df.head()
```

```python
## Preprocessing the data
## Drop unnecessary columns
df.drop(['RowNumber','CustomerId','Surname'],axis=1,inplace=True)
```

```python
df.head()
```

```python
## Encode   categorical variables
le_gender=LabelEncoder()
df['Gender']=le_gender.fit_transform(df['Gender'])
```

```python
ohe_geography=OneHotEncoder()
geography_encoded=ohe_geography.fit_transform(df[['Geography']]).toarray()
geography_encoded
```

```python
geo_encoded_df=pd.DataFrame(geography_encoded,columns=ohe_geography.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
## Combine ohe columns with df
df=pd.concat([df.drop(['Geography'], axis=1), geo_encoded_df], axis=1)
```

```python
df.head()
```

```python
## Save the encoders and sscaler
with open('le_gender.pkl','wb') as file:
    pickle.dump(le_gender,file)

with open('ohe_geography.pkl','wb') as file:
    pickle.dump(ohe_geography,file)
```
